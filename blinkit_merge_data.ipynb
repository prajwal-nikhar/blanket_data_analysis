```python
import pandas as pd
```

```python
# Load data files
customer_feedback = pd.read_csv('/Users/prajwalnikhar/Downloads/archive/blinkit_customer_feedback.csv')
customers = pd.read_csv('/Users/prajwalnikhar/Downloads/archive/blinkit_customers.csv')
marketing = pd.read_csv('/Users/prajwalnikhar/Downloads/archive/blinkit_marketing_performance.csv')
order_items = pd.read_csv('/Users/prajwalnikhar/Downloads/archive/blinkit_order_items.csv')
orders = pd.read_csv('/Users/prajwalnikhar/Downloads/archive/blinkit_orders.csv')
products = pd.read_csv('/Users/prajwalnikhar/Downloads/archive/blinkit_products.csv')
```

```python
customer_feedback.head(8)
```

```python
# Merge order_items with products on product_id
orders_items = pd.merge(order_items, orders, on='order_id', how='left')
```

```python
# Step 2: Add product information by merging products on 'product_id'
orders_items_products = pd.merge(orders_items, products, on='product_id', how='left')
```

```python
# Step 3: Combine feedback with customers on 'customer_id'
feedback_customers = pd.merge(customer_feedback, customers, on='customer_id', how='left')
```

```python
# Step 4: Merge the above feedback_customers with orders on 'order_id' to get order data aligned with feedback and customers
feedback_customers_orders = pd.merge(feedback_customers, orders, on='order_id', how='left', suffixes=('', '_order'))
```

```python
# Step 5: Combine feedback_customers_orders with order_items subset (removing overlapping columns from orders to avoid duplicates)
# Identify columns from order_items not in orders to avoid duplicate columns
order_items_extra_cols = order_items.columns.difference(orders.columns).tolist()
order_items_subset = order_items[['order_id'] + order_items_extra_cols]

final_dataset = pd.merge(feedback_customers_orders, order_items_subset, on='order_id', how='left')
```

```python
# Step 6: Add product details again for those order items (if needed)
final_dataset = pd.merge(final_dataset, products, on='product_id', how='left')
```

```python
# Save the merged dataset (for further analysis or Tableau)
blinkit_df = final_dataset.to_csv('/Users/prajwalnikhar/Downloads/blinkit_final_merged.csv', index=False, header=True)
print("File saved successfully at /Users/prajwalnikhar/Downloads/blinkit_final_merged.csv")
```
